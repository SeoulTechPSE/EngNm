**32[10].** $~$ 아래에 제시된 비선형 방정식 시스템의 해를 구하시오.


>$\begin{align}
y - x^3 -2x^2 &=-1 \\ 
 y+x^2&= 1 
\end{align}$



**Answer** $~$

````python
import numpy as np
from scipy import optimize

def f(x):
    return [x[1] -x[0]**3 -2*x[0]**2 +1, x[1] +x[0]**2 -1]

x = optimize.fsolve(f, [1, 1])
print(f'x1 = {x[0]}, x2 = {x[1]}')
````

**33[10].** $~$ Solve the following QP: 

>$\min_{x_1, x_2} 2x_1^2 +x_2^2 +x_1 x_2 + x_1 +x_2$ 

>subject to 

>$x_1 \geq 0$, $~$$x_2 \geq 0\text{ }$ and $\text{ }x_1 +x_2 =1$

**Answer** $~$

````python
%%capture
!pip install cvxopt
````

````python
import numpy as np
import cvxopt
````

````python
Q = 2*cvxopt.matrix([[2, 0.5], [0.5, 1]])
p = cvxopt.matrix([1.0, 1.0])
G = cvxopt.matrix([[-1.0, 0.0], [0.0,-1.0]])
h = cvxopt.matrix([0.0, 0.0])
A = cvxopt.matrix([1.0, 1.0], (1, 2))
b = cvxopt.matrix(1.0)

sol = cvxopt.solvers.qp(Q, p, G, h, A, b)
x = np.array(sol['x'])
print(f'x1 = {x[0]}, x2 = {x[1]}')
````

**35[10].** $~$ Solve the
Lokta-Volterra equations for the dynamics of a population of predator and prey animals (a classic example of
coupled ODEs). The equations are

$$
x'(t) = a x - b x y
$$

$$
y'(t) = c x y - d y
$$

where $x(t)$ is the number of prey animals and $y(t)$ is the number of predator animals, and the coefficients $a$, $b$, $c$, and $d$ describe the rates of the processes in the model


```
a, b, c, d = 0.4, 0.002, 0.001, 0.7
```

초기값 `[400, 600]`, 시간은 `[0, 80]` 범위에서 미분방정식을 풀고 그 결과를 그림으로 제시하시오.

**Answer** $~$

````python
import numpy as np
from scipy import integrate

a, b, c, d = 0.4, 0.002, 0.001, 0.7
def f(t, xy):
    x, y = xy
    return [a *x -b *x *y, c *x *y -d *y]

xy0 = [400, 600]
sol = integrate.solve_ivp(f, [0, 80], xy0, t_eval=np.linspace(0, 80, 400))

t = sol.t
xy_t = sol.y.T
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].plot(t, xy_t[:,0], 'r', label="Prey")
axes[0].plot(t, xy_t[:,1], 'b', label="Predator")
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Number of animals")
axes[0].legend()
axes[0].set_xlim(0, 80)
axes[0].set_ylim(0, 2000)
axes[0].tick_params(which='both', direction='in')

axes[1].plot(xy_t[:,0], xy_t[:,1], 'k')
axes[1].set_xlabel("Number of prey")
axes[1].set_ylabel("Number of predators")
axes[1].set_xlim(0, 2000)
axes[1].set_ylim(0, 800)
axes[1].tick_params(which='both', direction='in')
````
